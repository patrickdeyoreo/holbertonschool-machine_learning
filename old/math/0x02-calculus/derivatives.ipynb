```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import math
```

### `v = sin(u)`

```python
def v(u):
    return math.sin(u)
```

### `u = x^2`

```python
def u(x):
    return x ** 2
```

### `{x | x ∈ ℝ, 0 ≤ x ≤ 9}`

```python
X = np.linspace(0, 9, 5000)
X
```

### `{v | v ∈ ℝ}`

```python
V = np.array([v(u(x)) for x in X])
V
```

### `d(v)/dx = cos(x²) ⋅ 2x`

```python
def dvdx(x):
    return math.cos(x ** 2) * 2 * x
```

```python
dVdX = np.array([dvdx(x) for x in X])
```

### `v(x)` and `d(v)/dx`

```python
plt.title("v(x) and d(v)/dx")
plt.ylabel('amplitude')
plt.xlabel('x')

plt.scatter(X, V, color='b', marker='.', label='v(x)')
plt.scatter([math.pi], [v(u(math.pi))], color='orange', marker='o', edgecolor='white')
plt.scatter(X, dVdX, color='m', marker='.', label='d(v)/dx')
plt.scatter([math.pi], [dvdx(math.pi)], color='orange', marker='o', edgecolor='white')

plt.xlim([X.min(), X.max()])
plt.ylim(min([V.min(), dVdX.min()]), max([V.max(), dVdX.max()]))

plt.legend()
plt.show()
```
